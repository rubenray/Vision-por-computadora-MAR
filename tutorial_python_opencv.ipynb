## 3Ô∏è‚É£ Importar Librer√≠as

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

## 4Ô∏è‚É£ Subir y Cargar Imagen

```
from google.colab import files
uploaded = files.upload()
```

```
img = cv2.imread(list(uploaded.keys())[0])
cv2_imshow(img)
```

## 5Ô∏è‚É£ Procesamiento B√°sico

```
# Blanco y negro
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
# Bordes
edges = cv2.Canny(gray, 100, 200)
cv2_imshow(edges)
```

```
# Redimensionar
resized = cv2.resize(img, (300,300))
cv2_imshow(resized)
```

## 6Ô∏è‚É£ Guardar y Descargar Resultados

```
cv2.imwrite("resultado.png", edges)
files.download("resultado.png")
```
